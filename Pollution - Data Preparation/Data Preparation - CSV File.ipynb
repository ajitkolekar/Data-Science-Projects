```python
# Load necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read pollution_us_2000_2016.csv file, store data in data frame, and check size of the data frame using shape function

pollution_df = pd.read_csv("pollution_us_2000_2016.csv")
pollution_df.shape
```

```python
# Review first few records in the data frame using head() function

pollution_df.head()
```

```python
# After Reviewing the data, it appears that some data columns are not needed. Following columns will need to be deleted.

# Unnamed: 0 (column with index value of 0)
# State Code 
# County Code

del pollution_df['Unnamed: 0']
del pollution_df['State Code']
del pollution_df['County Code']
```

```python
# Confirm that the three columns are deleted

pollution_df.head()
```

```python
# Review the columns names to see if any column headings need to be changed

pollution_df.columns
```

```python
# Rename column names for max value and max hour to remove '1st' from the column names since that is irrelevant

pollution_df.rename(columns={'NO2 1st Max Value': 'NO2 Max Value', 
                            'NO2 1st Max Hour': 'NO2 Max Hour',
                            'O3 1st Max Value':'O3 Max Value',
                            'O3 1st Max Hour':'O3 Max Hour',
                            'SO2 1st Max Value':'SO2 Max Value',
                            'SO2 1st Max Hour':'SO2 Max Hour',
                            'CO 1st Max Value':'CO Max Value',
                            'CO 1st Max Hour':'CO Max Hour'}, inplace=True)
```

```python
# Confirm that the column names are updated

pollution_df.columns
```

```python
# The missing values are identified as null and can be checked by using isnull() function. Find missing values for each column
# and add them by using sum() function to determine how many missing values exist for each column.

pollution_df.isnull().sum()
```

```python
# The number of null values for SO2 AQI and CO AQI are almost 50% each compared to the original data set. AQI stands for 
# Air Quality Index, and it is an important measure for data analysis. Even after removing the rows with null values for 
# SO2 AQI and CO AQI, there would be sufficient records to conduct data analysis.

pollution_df = pollution_df[pd.notnull(pollution_df['SO2 AQI'])]
pollution_df = pollution_df[pd.notnull(pollution_df['CO AQI'])]
          
# Measure the size of the data frame after removing null values
pollution_df.shape
```

```python
# The nature of the data set and the data collected does not restrict from having duplicate values in any of the columns, 
# however, NO2 Units, O3 Units, SO2 Units, and CO Units columns have the same values for each row

# NO2 Units = "Parts per billion" 
# O3 Units = "Parts per million"
# SO2 Units = "Parts per billion"
# CO Units = "Parts per million"

# These four columns can be removed atter making the units consistent for O3 and CO measurements. They need to be multiplied 
# by 1000 to make them consistent with NO2 and SO2 measurements.

pollution_df['O3 Mean'] = pollution_df['O3 Mean']*1000
pollution_df['O3 Max Value'] = pollution_df['O3 Max Value']*1000
pollution_df['CO Mean'] = pollution_df['CO Mean']*1000
pollution_df['CO Max Value'] = pollution_df['CO Max Value']*1000

# Confirm that the changes are applied
pollution_df.head()
```

```python
# Delete columns for units.

# NO2 Units
# O3 Units 
# SO2 Units
# CO Units

del pollution_df['NO2 Units']
del pollution_df['O3 Units']
del pollution_df['SO2 Units']
del pollution_df['CO Units']

# Confirm that the columns are deleted
pollution_df.columns
```

```python
# The Site Num, Address, State, County, and City combinations are duplicated in the data frame. It would make sense to 
# normalize the data and create a separate data frame for these columns, and keep Site Num as the only column in the 
# primary data frame. However, the other data sources would me matched with County and City, it does not make sense to 
# separate the data and then join again. So, it is better to keep the data as is for now.
```

```python
# While there are significant outliers for the AQI values of each pollutant, have them for analysis is the purpose of 
# the project. Higher AQI value represents that the air is highly polluted. 

# Following are the measures for AQI values (as per https://www.airnow.gov/aqi/aqi-basics/):
# AQI between 0 and 50: Air quality satisfactory
# AQI between 51 to 100: Air quality acceptable, risky for sensitive people
# AQI between 101 to 150: Sensitive people may experience health effects, general public less likely to be affected
# AQI between 151 to 200: Sensitive people may experience serious health effects, general public may be affected
# AQI between 201 to 300: Health alert: The risk of health effects is increased for everyone
# AQI 301 or higher: Health warning of emergency conditions: everyone is more likely to be affected

# As I plan to compare the AQI values with the population, it makes sense to not remove the outliers and use them for comparison
```

```python
# Most of the columns have numeric values. State, County, and City values are consistent and are stored with title case
# (first character of each word to uppercase and remaining to lowercase). The Address column values need to be converted 
# to title case. 

pollution_df['Address'] = pollution_df['Address'].str.title()

# Check if the changes are applied using head() function
pollution_df.head()
```
